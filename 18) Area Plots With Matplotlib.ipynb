```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a quick dataFrame
my_df = pd.DataFrame(randn(25,4),columns = ["Mon","Tues","Wed","Thur"])
my_df
```

```python
# Area plot
my_df.plot(kind="area",stacked=False)
# We used stacked false because there are negative and positive values
```

```python
my_df.abs().plot(kind="area")
```

```python
# Shading
my_df.abs().plot(kind="area",alpha=0.5)
```

```python
# Title
my_df.abs().plot(kind="area",alpha=0.5, title=" My Awesome Are Plot")
```

```python
# Legends
# Title
my_df.abs().plot(kind="area",alpha=0.5, title=" My Awesome Are Plot", legend=False)
```

```python
# Plot only one column
# Title
my_df["Mon"].abs().plot(kind="area",alpha=0.5, title=" My Awesome Are Plot")
```

```python
# Plot two columns
my_df[["Mon","Tues"]].abs().plot(kind="area",alpha=0.5, title=" My Awesome Are Plot",legend = False)
```

```python
# Add table
my_df[["Mon","Tues"]].abs().plot(kind="area",table=True)

```
