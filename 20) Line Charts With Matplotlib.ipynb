```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a quick dataFrame
my_df = pd.DataFrame(randn(25,4),columns = ["Mon","Tues","Wed","Thur"])
my_df
```

```python
# Line plot
my_df.plot(kind="line")
```

```python
# Absolute line plot
my_df.abs().plot(kind="line")
```

```python
# Line width
my_df.plot(kind="line",lw=5)
```

```python
# Chart just one column
my_df["Mon"].abs().plot(kind="line")
```

```python
# Line more than one
my_df[["Mon","Tues"]].abs().plot(kind="line")
```

```python
# Size
my_df[["Mon","Tues"]].abs().plot(kind="line",figsize=(15,5))
```

```python
# Title
my_df.abs().plot(kind="line",title="My Line Chart")
```

```python
# Shading
my_df.abs().plot(kind="line",alpha=0.3)
```

```python
# Legend
my_df.abs().plot(kind="line",legend=False)
```

```python
# Second way chart
my_df.abs().plot.line(lw=0.5)
```
