```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a quick dataFrame
my_df = pd.DataFrame(randn(50,4),columns = ["Mon","Tues","Wed","Thur"])
my_df
```

```python
# Density and Kernel Density Estimation(KDE) plots with scipy 
my_df.plot(kind="kde")
# density = yoğunluk
```

```python
# Second way
my_df.plot.kde()
```

```python
# Density plot
my_df.plot.density()
my_df.plot(kind="density")
```

```python
# Change stuff
my_df[["Mon","Tues"]].plot.density(alpha=0.3,grid=True,lw=5,title="My Density Chart")
```
