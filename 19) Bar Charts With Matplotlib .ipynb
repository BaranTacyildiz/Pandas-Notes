```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a quick dataFrame
my_df = pd.DataFrame(randn(25,4),columns = ["Mon","Tues","Wed","Thur"])
my_df
```

```python
# Bar Plot
# my_df.plot(kind="bar")
my_df.abs().plot(kind="bar")
```

```python
# Bar plot stocked
my_df.abs().plot(kind="bar",stacked=True)
```

```python
# Shading
my_df.abs().plot(kind="bar",stacked=True,alpha=0.3)
```

```python
# Title
my_df.abs().plot(kind="bar",stacked=True,title="My Bar Chart")
```

```python
# Legend
my_df.abs().plot(kind="bar",stacked=True,title="My Bar Chart",legend=False)
```

```python
# Second Way
my_df.abs().plot.bar()
```

```python
# Second way stacked
my_df.abs().plot.bar(stacked=True)
```

```python
# Second way Shading
my_df.abs().plot.bar(alpha=0.3)
```

```python
# Second way title
my_df.abs().plot.bar(title="My Bar Chart")
```

```python
# Second way legend
my_df.abs().plot.bar(legend=False)
```

```python
# Grid Lines
my_df.abs().plot.bar(stacked=True,grid=True)
```
