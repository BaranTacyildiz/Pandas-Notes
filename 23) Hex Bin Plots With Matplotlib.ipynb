```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a quick dataFrame
my_df = pd.DataFrame(randn(50,4),columns = ["Mon","Tues","Wed","Thur"])
my_df
```

```python
# Create hex plot
my_df.plot(kind="hexbin",x="Mon",y="Tues")
```

```python
# Hex plots gridsize
my_df.plot(kind="hexbin",x="Mon",y="Tues",gridsize=10)
```

```python
# Color
my_df.plot(kind="hexbin",x="Mon",y="Tues",gridsize=10,cmap="Reds")
```

```python
# Second way
my_df.plot.hexbin(x="Mon",y="Tues",gridsize=10)
```

```python
my_df.plot.hexbin(x="Mon",y="Tues",C="Wed",gridsize=10)
```

```python
# Title
# Second way
my_df.plot.hexbin(x="Mon",y="Tues",gridsize=10,title="My Hex Bin Chart")
```

```python
# Grid
# Second way
my_df.plot.hexbin(x="Mon",y="Tues",gridsize=10,grid=True)
```
