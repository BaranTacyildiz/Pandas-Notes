```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a quick dataFrame
my_df = pd.DataFrame(randn(100,4),columns = ["Mon","Tues","Wed","Thur"])
my_df
```

```python
# Histogram
my_df["Wed"].hist()
```

```python
# Histogram with smaller bins,50 ,20 (bin = kutu)
my_df["Wed"].hist(bins=20)
```

```python
# Another way without grid bars
my_df["Wed"].plot(kind="hist",bins=20)
```

```python
# plot(legend=)
my_df["Wed"].plot(kind="hist",bins=20,legend = True)
```
