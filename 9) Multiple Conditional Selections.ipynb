```python
import pandas as pd
import numpy as np
my_df = pd.read_csv("all_years.csv")
my_df
```

```python
# Multiple Conditional AND
my_df[(my_df["Gender"]=="MALE") & (my_df["Year"]>2010)]
# Be careful its & not "and"
```

```python
# Get Length
len(my_df[(my_df["Gender"]=="MALE") & (my_df["Year"]>2010)])

```

```python
# Multiple Conditional OR
my_df[(my_df["Gender"]=="MALE") | (my_df["Year"]>2010)]
```

```python
# Get length
len(my_df[(my_df["Gender"]=="MALE") | (my_df["Year"]>2010)])
```

```python
# Return Spesific Column
my_df[(my_df["Gender"]=="MALE") & (my_df["Year"]>2010)]["Dog_name"]
```
