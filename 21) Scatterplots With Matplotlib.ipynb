```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a quick dataFrame
my_df = pd.DataFrame(randn(50,4),columns = ["Mon","Tues","Wed","Thur"])
my_df
```

```python
# Create a scatter plot
my_df.plot(kind="scatter", x="Mon",y="Tues")
# U must to identify x and y coordinates to run scatter plot
```

```python
my_df.plot(kind="scatter", x="Mon",y="Tues",c="Wed")
```

```python
# Use colors (cmap=color map) (for colors search = https://matplotlib.org/stable/users/explain/colors/index.html#tutorials-colors)
my_df.plot(kind="scatter", x="Mon",y="Tues",c="Wed",cmap="magma")
```

```python
#By size easier to read with alpha
my_df.plot(kind="scatter", x="Mon",y="Tues",s=my_df["Wed"]*200,alpha=0.6)
```

```python
# Title
my_df.plot(kind="scatter", x="Mon",y="Tues",c="Wed",title="My Scatter Plot")
```
